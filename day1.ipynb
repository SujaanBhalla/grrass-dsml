```python
# logical operators practice

num1 = 10
num2 = 20

print("AND with numbers")
print(num1 and num2)

print("OR with numbers")
print(num1 or num2)

print("NOT with number")
print(not num1)

print("----------------")

num1 = "hello"
num2 = ""

print("OR with strings")
print(num1 or num2)
print(num2 or num1)

print("NOT with string")
print(not num1)
print(not num2)

print("----------------")

num1 = []

print("empty list")
print(num1)

print("NOT on empty list")
print(not num1)

print("----------------")

num1 = [10,20,30]

print("list with values")
print(num1)

print("NOT on list with values")
print(not num1)

print("----------------")

num1 = {}

print("empty dictionary")
print(num1)

print("NOT on empty dictionary")
print(not num1)

print("----------------")

num1 = {"name":"sujaan"}

print("dictionary with value")
print(num1)

print("NOT on dictionary with value")
print(not num1)

print("----------------")

a = 1
b = 0

print("AND")
print(a and b)

print("OR")
print(a or b)

print("NOT")
print(not a)

print("----------------")

print("XOR")
print(a ^ b)
```

```python
# example 4

class student:

    def __init__(self, *args):
        self.data = args

    def users(self):
      print(self.args["dheeraj"])
      print(self.args["harsh"])


    def details(self):
       print(self.args["address"])
       print(self.args["college"])
       print(self.args["loc"])


s = student(
    {
        "dheeraj":"kunal",
        "harsh":"praveen",
        "address":"kukas",
        "college":"arya",
        "loc":"jaipur"
    })
s.users()
s.details()
```
